# Bike Sharing Assignement

### Problem Statements

A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. Many bike share systems allow people to borrow a bike from a "dock" which is usually computer-controlled wherein the user enters the payment information, and the system unlocks it. 

The company wants to know:

- Which variables are significant in predicting the demand for shared bikes.
- How well those variables describe the bike demands

The steps we will follow in this exercise are as follows:

1. Reading, understanding and visualising the data
2. Preparing the data for medelling (train-test split, rescaling etc.)
3. Training the model
4. Residual Analysis
5. Predictions and evaluation on the test set

### Step 1: Reading and understaning the Data

import NumPy and Pandas and read the Bike Sharing Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# read the data

bike_assignment = pd.read_csv(r'C:\Software\Data\Gaurav Certificates\PGD - Data Science Notes\Machine Learning\day.csv')
bike_assignment.head()
```

```python
# drop the above variables ['instant', 'dteday', 'temp', 'casual', 'registered'] 
bike_assignments = bike_assignment.drop(['instant', 'dteday', 'temp', 'casual', 'registered'], axis=1)
bike_assignments.head()
```

```python
bike_assignments.shape
```

```python
bike_assignments.info()
```

```python
# Replace Categorical & Binary variables [weathersit, season, mnth, weekday], [workingday, yr, holiday]
bike_assignments['season'].replace({1:'spring', 2:'summer', 3:'fall', 4:'winter'}, inplace =True)
bike_assignments['weathersit'].replace({1:'Clear', 2:'Mist + Cloudy', 3:'Light Snow', 4:'Heavy Rain + Ice Pallets'}, inplace =True)
bike_assignments['mnth'].replace({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}, inplace =True)
bike_assignments['weekday'].replace({0:'Sun', 1:'Mon', 2:'Tue', 3:'Wed', 4:'Thu', 5:'Fri', 6:'Sat'}, inplace =True)
bike_assignments.head()
```

```python
bike_assignments.describe()
```

```python
# visualising a numerical veriables
sns.pairplot(bike_assignments)
plt.show()
```

```python
# creating dummy variables for season, mnth, weekday, & weathersit. Next drop the redundant dummy variables
dummy_season=pd.get_dummies(bike_assignments['season'],drop_first=True)
dummy_mnth=pd.get_dummies(bike_assignments['mnth'],drop_first=True)
dummy_weekday=pd.get_dummies(bike_assignments['weekday'],drop_first=True)
dummy_weathersit=pd.get_dummies(bike_assignments['weathersit'],drop_first=True)

#concat the dummy dataframe with base data
bike_assignments=pd.concat([bike_assignments,dummy_season],axis=1)
bike_assignments=pd.concat([bike_assignments,dummy_mnth],axis=1)
bike_assignments=pd.concat([bike_assignments,dummy_weekday],axis=1)
bike_assignments=pd.concat([bike_assignments,dummy_weathersit],axis=1)
bike_assignments.info()
```

```python
# dropping the initial categorical variable
bike_assignments=bike_assignments.drop('season',axis=1)
bike_assignments=bike_assignments.drop('mnth',axis=1)
bike_assignments=bike_assignments.drop('weekday',axis=1)
bike_assignments=bike_assignments.drop('weathersit',axis=1)
bike_assignments.info()
```

### Step 2: Preparing the Data for Modelling
 - Splitting into train and test
 - Rescaling of variables

### Splitting into train and test

```python
df_train, df_test = train_test_split(bike_assignments, train_size=0.7, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

### Rescaling the Features
1. Min-Max scaling (normalisation): Between 0 and 1
2. Standardisation(mean-0, sigma-1)

This time, we will use MinMax scaling

```python
# 1. Instantiate an object
scaler = MinMaxScaler()

# create a list of numeric vars
num_vars = ['atemp', 'hum','windspeed', 'cnt']

# 2. Fit on data
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```
